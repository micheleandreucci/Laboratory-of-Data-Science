# PLAYER

```python
# We use the csv obtained from the splitting of the tennis csv file
df_player = pd.read_csv('player.csv')
```

```python
df_player['match_id'] = df_player['tourney_id'] + df_player['match_num'].astype(str)
```

```python
df_player['match_id'] = df_player['match_id'].drop_duplicates()
```

```python
df_player = df_player[df_player['match_id'].notna()]
```

```python
df_player
```

```python
# We create our final columns and we delete the ones we obtained them from

player_id = df_player['winner_id'].append(df_player['loser_id'], ignore_index=True).rename('player_id')
name = df_player['winner_name'].append(df_player['loser_name'], ignore_index=True).rename('name')
hand = df_player['winner_hand'].append(df_player['loser_hand'], ignore_index=True).rename('hand')
ht = df_player['winner_ht'].append(df_player['loser_ht'], ignore_index=True).rename('ht')
country_id = df_player['winner_ioc'].append(df_player['loser_ioc'], ignore_index=True).rename('country_id')
age = df_player['winner_age'].append(df_player['loser_age'], ignore_index=True).rename('age')
tourney_date =  df_player['tourney_date'].append(df_player['tourney_date'], ignore_index=True).rename('tourney_date1')

df1 = df_player.join([player_id,name,hand,ht,country_id,age,tourney_date], how='outer')
```

```python
del df1['tourney_date']
df1 = df1.rename(columns={'tourney_date1':'tourney_date'})
```

```python
# We delete useless columns

del df1['winner_id']
del df1['loser_id']
del df1['winner_name']
del df1['loser_name']
del df1['winner_ht']
del df1['winner_hand']
del df1['winner_ioc']
del df1['loser_ioc']
del df1['winner_age']
```

```python
del df1['loser_ht']
del df1['loser_hand']
del df1['loser_age']
```

```python
df1
```

```python
len(df1['player_id'].unique())
```

```python
len(df1['name'].unique())
```

```python
# We save the updated csv file with the new columns

df1.to_csv('Giocatori.csv')
```

```python
date['date_id'] = date['date_id'].drop_duplicates()
date = date[date['date_id'].notna()]
date.reset_index(drop=True, inplace=True)
```

```python
import pandas as pd
match=pd.read_csv('match.csv',sep=',')
#del match['Unnamed: 0']
```

```python
match.head()
```

```python
match = match[match['winner_id'] != match['loser_id']]
```

```python
match['match_id'] = match['tourney_id'] + match['match_num'].astype(str)
```

```python
match['match_id'].drop_duplicates()
```

```python
match['match_id'] = match['match_id'].drop_duplicates()
match = match[match['match_id'].notna()]
```

```python
match.isnull().sum()
```

```python
match['match_id']=match['tourney_id'] + match['match_num'].astype(str)
```

```python
columns = ['tourney_id', 'match_id','winner_id', 'loser_id', 
       'score', 'best_of', 'round', 'minutes', 'w_ace', 'w_df', 'w_svpt',
       'w_1stIn', 'w_1stWon', 'w_2ndWon', 'w_SvGms', 'w_bpSaved', 'w_bpFaced',
       'l_ace', 'l_df', 'l_svpt', 'l_1stIn', 'l_1stWon', 'l_2ndWon', 'l_SvGms',
       'l_bpSaved', 'l_bpFaced', 'winner_rank', 'winner_rank_points',
       'loser_rank', 'loser_rank_points']
match = match.reindex(columns=columns)
```

```python
match["score"] = match["score"].fillna(match["score"].mode()[0])
```

```python
import numpy as np
match["minutes"] = np.floor(match["minutes"].fillna(round(match["minutes"].mean())))
```

```python
match['loser_rank_points'] = match['loser_rank_points'].fillna(round(match['loser_rank_points'].mean()))
```

```python
match['winner_rank_points'] = match['winner_rank_points'].fillna(round(match['winner_rank_points'].mean()))
```

```python
match['winner_rank'] = match['winner_rank'].fillna(round(match['winner_rank'].mean()))
```

```python
match['loser_rank'] = match['loser_rank'].fillna(round(match['loser_rank'].mean()))
```

```python
match["winner_rank_points"][(match["winner_id"] == list(match["winner_id"]))] = (match["winner_rank_points"].fillna(match["winner_rank_points"][(match["winner_id"]==list(match["winner_id"]))].mode()[0]))
```

```python
match["loser_rank_points"][(match["loser_id"] == list(match["loser_id"]))] = (match["loser_rank_points"].fillna(match["loser_rank_points"][(match["loser_id"]==list(match["loser_id"]))].mode()[0]))
```

```python
import pandas as pd
df = pd.read_csv('player.csv',sep=',')
#del df['Unnamed: 0']
```

```python
df1[df1['country_ioc']=='COD']
```

```python
continent = df.merge(df1,how='left',on="country_ioc")
```

```python
continent[continent['country_ioc']=='COD']
```

```python
continent['country_name'] = continent['country_name'].fillna('Unknown')
continent['continent'] = continent['continent'].fillna('Unknown')
```

```python
continent[continent['country_ioc']=='COD']
```
